## Segformer

```python
from mmseg.apis import inference_segmentor, init_segmentor
import mmcv
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# config, checkpoint 파일 위치 지정 (경로는 자신의 환경에 맞게)
config_file = 'configs/segformer/segformer_mit-b5_512x512_160k_coco-stuff164k.py'
checkpoint_file = 'segformer_mit-b5_512x512_160k_coco-stuff164k_20220208_231808-03660c89.pth'
```

```python
# 모델 초기화
model = init_segmentor(config_file, checkpoint_file, device='cuda:0')  # or 'cpu'
```

```python
# 테스트 이미지 지정
img_path = 'cat_on_sofa.jpg'  # (파일명/경로를 자신의 이미지로 변경)
```

```python
# 추론
result = inference_segmentor(model, img_path)
```

```python
# 결과 시각화 (파일 저장/직접 보기 둘 다 가능)
out_file = 'segformer_coco_result.png'
model.show_result(img_path, result, show=True, opacity=0.7, out_file=out_file)

# Jupyter에서 바로 시각화 (원하면)
img = Image.open(out_file)
plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```
